```python
import yfinance as yf
import plotly.graph_objects as go

# Download Tesla stock data for the last year
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="1y")

# Get Tesla's quarterly revenue data (from income statement)
income_statement = tesla.quarterly_financials
revenue_data = income_statement.loc['Total Revenue']

# Convert quarterly revenue data into a format for plotting
revenue_data = revenue_data.sort_index(ascending=True)  # Ensure the data is sorted chronologically

# Create the figure
fig = go.Figure()

# Add Tesla stock price (line plot)
fig.add_trace(go.Scatter(x=tesla_data.index, y=tesla_data['Close'], mode='lines', name='Tesla Stock Price', line=dict(color='blue')))

# Add Tesla revenue (bar plot)
fig.add_trace(go.Bar(x=revenue_data.index, y=revenue_data.values / 1e9, name='Tesla Revenue (in Billion $)', marker=dict(color='red')))

# Customize the layout
fig.update_layout(
    title="Tesla Stock Price vs Revenue",
    xaxis_title="Date",
    yaxis_title="Stock Price (USD)",
    yaxis2=dict(
        title="Revenue (Billion USD)",
        overlaying="y",
        side="right"
    ),
    template="plotly_dark",
    barmode="group",  # Group bar and line chart
)

# Show the interactive plot
fig.show()

```

```python
import yfinance as yf
import plotly.graph_objects as go

# Download GameStop stock data for the last year
gamestop = yf.Ticker("GME")
gamestop_data = gamestop.history(period="1y")

# Get GameStop's quarterly revenue data (from income statement)
income_statement_gme = gamestop.quarterly_financials
revenue_data_gme = income_statement_gme.loc['Total Revenue']

# Convert quarterly revenue data into a format for plotting
revenue_data_gme = revenue_data_gme.sort_index(ascending=True)  # Ensure the data is sorted chronologically

# Create the figure
fig_gme = go.Figure()

# Add GameStop stock price (line plot)
fig_gme.add_trace(go.Scatter(x=gamestop_data.index, y=gamestop_data['Close'], mode='lines', name='GameStop Stock Price', line=dict(color='green')))

# Add GameStop revenue (bar plot)
fig_gme.add_trace(go.Bar(x=revenue_data_gme.index, y=revenue_data_gme.values / 1e9, name='GameStop Revenue (in Billion $)', marker=dict(color='orange')))

# Customize the layout
fig_gme.update_layout(
    title="GameStop Stock Price vs Revenue",
    xaxis_title="Date",
    yaxis_title="Stock Price (USD)",
    yaxis2=dict(
        title="Revenue (Billion USD)",
        overlaying="y",
        side="right"
    ),
    template="plotly_dark",
    barmode="group",  # Group bar and line chart
)

# Show the interactive plot
fig_gme.show()
```
